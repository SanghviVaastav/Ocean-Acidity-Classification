```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. FINAL BENCHMARKING CHART
models = ['Logistic Regression', 'Random Forest', 'XGBoost']
f1_scores = [0.9937, 0.9932, 0.9958]

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
bars = plt.bar(models, f1_scores, color=['red', 'orange', 'green'], alpha=0.8)
plt.ylim(0.99, 1.0)
plt.title(' Metric Benchmarking: Final Model Comparison', fontsize=14, fontweight='bold')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.0001, f"{bar.get_height():.4f}", ha='center')
plt.savefig('_benchmarking.png')
plt.show()

# 2. CONFUSION MATRIX (XGBoost)
cm_data = np.array([[1569, 0, 0], [0, 19397, 0], [0, 0, 9034]])
classes = ['Critical', 'Safe', 'Vulnerable']
plt.figure(figsize=(8, 7))
sns.heatmap(cm_data, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.title(' Confusion Matrix Analysis: XGBoost', fontsize=14, fontweight='bold')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig('_confusion_matrix.png')
plt.show()

# 3. INTERPRETABILITY (Feature Importance)
features = ['fCO2_per_SST', 'SST_sal_interaction', 'SST', 'abs_lat', 'lat', 'lon', 'shipping_intensity']
importance = [0.681, 0.063, 0.057, 0.037, 0.029, 0.021, 0.015]
plt.figure(figsize=(10, 6))
plt.barh(features, importance, color='#577590')
plt.gca().invert_yaxis()
plt.title(' Interpretability Drivers: Top Features', fontsize=14, fontweight='bold')
plt.savefig('_interpretability.png')
plt.show()
```
